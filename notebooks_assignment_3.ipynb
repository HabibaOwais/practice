```python
# Question: 01
number_1 = int(input("Enter number: "))
number_2 = int(input("Enter number: "))
sign = input("Enter any operator")

operators = {
    "+": number_1 + number_2, 
    "-": number_1 - number_2,
    "*": number_1 * number_2, 
    "/": number_1 / number_2,
    "**": number_1 ** number_2
}
if sign in operators:
    print("Result:", operators[sign])
else:
    print("Invalid operator")
```
